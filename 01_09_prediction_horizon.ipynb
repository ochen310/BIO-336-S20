# Prediction Horizon Exercise

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nsteps = 100              
ntraces = 10
x = np.zeros(shape=(nsteps,ntraces))    
wiggle = 0.0001

for r in np.arange(1,5): 
    for itrace in range(ntraces):
        x[0,itrace] = 0.6+np.random.randn(1)*wiggle
        for istep in range(1,nsteps):
            x[istep,itrace] = r*x[istep-1,itrace]*(1-x[istep-1,itrace])

figb = plt.figure(figsize=(14,7))
plt.plot(x)
plt.xlabel('Time');
plt.ylabel('Variability of Trajectories');
```

I set r equal to `np.arange(1,5)` so that the values of r would vary between 1 and 5. The trajectories begin to diverge at around `t = 5` so I'd that one could only confidently make predictions prior to that period of time. 

# Prediction Horizon Exercise Solution

```python
nsteps = 100              
ntraces = 10
x = np.zeros(shape=(nsteps,ntraces))    
wiggle = 0.0001

for r in np.arange(1,5): 
    for itrace in range(ntraces):
        x[0,itrace] = 0.6+np.random.randn(1)*wiggle
        for istep in range(1,nsteps):
            x[istep,itrace] = r*x[istep-1,itrace]*(1-x[istep-1,itrace])

figb = plt.figure(figsize=(14,7))
plt.plot(np.std(x,axis=1))
plt.xlabel('Time');
plt.ylabel('Variability of Trajectories');
```

```python
nsteps = 100              
ntrials = 100
x = np.zeros(shape=(nsteps,2))    
wiggle = 0.0001 # difference in starting values
diff = np.zeros(shape=(nsteps, ntrials)) # absolute difference in trajectories
fig = plt.figure(figsize=(14,7))

for r in np.arange(3.6,4,0.1): 
    for itrials in range(ntrials):
        starting_point = np.random.uniform(low=0,high=1-wiggle) + np.array([0,wiggle])
        for ipair in range(2): 
            x[0, ipair]= starting_point[ipair]
            for istep in range(1,nsteps):
                x[istep,ipair] = r*x[istep-1,ipair]*(1-x[istep-1,ipair])
        diff[:, itrials] = np.abs(x[:,1]-x[:,0])  
    plt.plot(np.mean(diff, axis=1));
```
