```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
p = np.zeros(n)
p[0] = 1
p[1] = 1
for i in range(2,n):
    p[i] = p[i-1]*(1-(i-1)/365)
x = np.arange(start=2, stop=100, step=1);
plt.plot(1-p);
plt.xlabel('Number of people in group');
plt.ylabel('Shared birthdays in group of people');
```
