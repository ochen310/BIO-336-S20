```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 365
i = np.arange(2,n)
def p(i):
    print(i*(1-(1/365)))
    
print(p(i))
```

```python
plt.figure(figsize=(5,5))
plt.plot(i,i*(1-(1/365)));
plt.xlabel('Number of people in a group');
plt.ylabel('Shared birthdays in a group of people');
```

```python
n = 365
x = (i-(i-1))/2
i = np.arange(2,n)
def p(i):
    print(1-(i/365)**x)
    
print(p(i))
```

```python
plt.plot(i,1-(i/365**x));
plt.xlabel('Number of people in a group');
plt.ylabel('Shared birthdays in a group of people');
```
