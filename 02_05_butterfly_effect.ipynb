```python
import numpy as np
import matplotlib.pyplot as plt
```

# Exercise 1: Compare two trajectories visually

```python
nsteps = 100
x = np.zeros(nsteps)
starting_points = [0.67,0.66]
fig = plt.figure(figsize=(10,7))
r = 3.9
for x[0] in starting_points:
    for i in range(1,nsteps):
        x[i] = r*x[i-1]*(1-x[i-1])
        
    plt.plot(x);
    
```
