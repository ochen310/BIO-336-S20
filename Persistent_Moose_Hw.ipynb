```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
%matplotlib notebook 

n_animals = 100
n_steps = 100
speed = 0.01

fig = plt.figure(figsize=(5,5));
plt.show()
plt.xlim((0,1))
plt.ylim((0,1))

for i in range(n_animals):
    v = np.random.rand(n_animals)
    for t in range(n_steps):
        x = np.ones(n_animals)*0.5 
        y = np.ones(n_animals)*0.5
        x+=v*x
        y+=v*y
        
    plt.clf()
    plt.plot(x,y,'k.',markersize=2);
    fig.canvas.draw() 
    time.sleep(0.01) 
```

I was able to get my moose to run in straight lines, but they all seem to be going in the same direction even though I set randomly distributed values for velocity.
