```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
%matplotlib notebook 

n_animals = 100
n_steps = 100
speed = 0.01

fig = plt.figure(figsize=(5,5));
plt.show()
plt.xlim((0,1))
plt.ylim((0,1))

for i in range(n_animals):
    v = np.random.rand(n_animals)
    for t in range(n_steps):
        x = np.ones(n_animals)*0.5 
        y = np.ones(n_animals)*0.5
        x+=v*x
        y+=v*y
        
    plt.clf()
    plt.plot(x,y,'k.',markersize=2);
    fig.canvas.draw() 
    time.sleep(0.01) 
```

```python
%matplotlib notebook

n_animals = 100
n_steps = 1000
vmin=0.01
vmax=0.05

x = np.ones(n_animals)*0.5
y = np.ones(n_animals)*0.5
vx = np.random.uniform(low=vmin,high=vmax,size=n_animals)*(2*np.random.randint(low=0,high=2,size=n_animals)-1)
vy = np.random.uniform(low=vmin,high=vmax,size=n_animals)*(2*np.random.randint(low=0,high=2,size=n_animals)-1)

# Plotting
fig = plt.figure(figsize=(3,3))
ax = plt.subplot(111)
plt.xlim((0,1));
plt.ylim((0,1));
fig.show()
fig.canvas.draw()

for t in range(n_steps): # moving elks around
    x += vx
    y += vy
    
    vx = np.where(np.logical_or(x>1,x<0),-vx,vx)
    vy = np.where(np.logical_or(y>1,y<0),-vy,vy)
    
    ax.clear()
    plt.plot(x,y,'k.',markersize=2);
    plt.xlim((0,1));
    plt.ylim((0,1));
    fig.canvas.draw() # <--------------- The main command to make it move
    #time.sleep(0.01)
```
