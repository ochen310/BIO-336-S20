# Growth

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nsteps = 100 # number of steps in sim
x = np.zeros(nsteps) # number of animals at each step
x[0] = 1 # the initial number of animals
r = 0.9 # rate of population change (change in growth and death rates)
for i in range(1,nsteps):
    x[i] = x[i-1]*r   

plt.plot(x);
```

```python
# Logistic growth model
nsteps = 100 # number of steps in sim
x = np.zeros(nsteps) # number of animals at each step
x[0] = 0.8 # the initial population size
r = 3.44949 # rate of population change (change in growth and death rates)
for i in range(1,nsteps):
    x[i] = x[i-1]*r*(1-x[i-1])
    
plt.plot(x);
plt.ylim(0,1) 
# The solution changes from single to double oscillation at r = 3.44949
```
